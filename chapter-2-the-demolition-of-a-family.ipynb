```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Data abd prepare it

```python
data_1 = pd.read_csv("/kaggle/input/demolition-and-displacement-in-palestine/Demolitions in the Occupied Palestinian Territory.csv")
```

```python
data_1.Total = data_1.Total.apply(lambda x:x.replace(",", ""))
data_1.Administrative = data_1.Administrative.apply(lambda x:x.replace(",", ""))
data_1.Military = data_1.Military.apply(lambda x:x.replace(",", ""))
data_1.Undefined = data_1.Undefined.apply(lambda x:x.replace(",", ""))

data_1 = data_1.astype(int)
```

```python
data_1
```

# Analyse

```python
print(f"The largest number of Military Demolitions in a single year is {data_1.Military.max()},", 
      f"which took place in {data_1[data_1.Military == data_1.Military.max()].Year.values[0]}")

print(f"The largest number of Punitive Demolitions in a single year is {data_1.Punitive.max()},", 
      f"which took place in {data_1[data_1.Punitive == data_1.Punitive.max()].Year.values[0]}")

print(f"The largest number of Administrative Demolitions in a single year is {data_1.Administrative.max()},", 
      f"which took place in {data_1[data_1.Administrative == data_1.Administrative.max()].Year.values[0]}")

print(f"The largest number of Undefined Demolitions in a single year is {data_1.Undefined.max()},", 
      f"which took place in {data_1[data_1.Undefined == data_1.Undefined.max()].Year.values[0]}")
```

```python
ax = plt.axes()
ax.set_title("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) since 1967")
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Number of Buildings", fontsize=15)
ax.legend("Total Number of Buildings Demolished")
data_1.plot.bar(x="Year", y="Total", ax=ax, figsize=(15, 10))
# plt.savefig("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) since 1967.png")
plt.show()
```

```python
ax = plt.axes()
ax.set_title("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) 1987-1993 (First Intifada)")
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Number of Buildings", fontsize=15)
ax.legend("Total Number of Buildings Demolished")
data_1.iloc[20:27, :-1].plot.bar(x="Year", ax=ax, figsize=(15, 10))
# plt.savefig("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) 1987-1993 (First Intifada).png")
plt.show()
```

```python
ax = plt.axes()
ax.set_title("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) 2000-2005 (Second Intifada)")
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Number of Buildings", fontsize=15)
ax.legend("Total Number of Buildings Demolished")
data_1.iloc[33:39, :-1].plot.bar(x="Year", ax=ax, figsize=(15, 10))
# plt.savefig("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) 2000-2005 (Second Intifada).png")
plt.show()
```

```python
ax = plt.axes()
ax.set_title("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) 2023-2025 (Palestinian-Israeli War)")
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Number of Buildings", fontsize=15)
ax.legend("Total Number of Buildings Demolished")
data_1.iloc[-3:, :-1].plot.bar(x="Year", ax=ax, figsize=(15, 10))
# plt.savefig("Buildings Demolished or Seized in Occupied Palestine (West Bank, East Jerusalem, and Gaza) 2023-2025 (Palestinian-Israeli War).png")
plt.show()
```
